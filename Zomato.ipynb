```python
zmt = zmt.drop(["url","address","phone","dish_liked","menu_item","reviews_list"],axis=1)
zmt.head()
```

```python
zmt.info()
```

```python
#dropping duplicates
zmt.drop_duplicates(inplace=True)
zmt.shape
```

```python
#cleaning rate column
zmt["rate"].unique()
```

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value = str(value).split("/")
        value = value[0]
        return float(value)
    
zmt["rate"] = zmt["rate"].apply(handlerate)
zmt["rate"].head()
    
```

```python
zmt.rate.isnull().sum()
```

```python
zmt["rate"].fillna(zmt["rate"].mean(), inplace=True)
```

```python
zmt.rate.isnull().sum()
```

```python
zmt.info()
```

```python
#dropping null values

zmt.dropna(inplace=True)
zmt.head()
```

```python
zmt.info()
```

```python
#renaming columns

zmt.rename(columns={"approx_cost(for two people)":"Cost2plates","listed_in(type)":"Type"}, inplace=True)
zmt.head()
```

```python
zmt["location"].unique()
```

```python
zmt["listed_in(city)"].unique()
```

```python
# keeping only one location
```

```python
zmt = zmt.drop(["listed_in(city)"],axis=1)
```

```python
zmt.head()
```

```python
zmt["Cost2plates"].unique()
```

```python
# removing ',' from Cost2plates column

def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
zmt['Cost2plates'] = zmt['Cost2plates'].apply(handlecomma)
zmt['Cost2plates'].unique()
```

```python
zmt.head()
```

```python
# cleaning rest_type column

zmt["rest_type"].value_counts()
```

```python
rest_type = zmt["rest_type"].value_counts(ascending = False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type<1000]
rest_type_lessthan1000
```

```python
# replacing lessthan 1000 to others

def handle_rest_type(value):
    if value in rest_type_lessthan1000:
        return 'others'
    else:
        return value
    
zmt['rest_type'] = zmt['rest_type'].apply(handle_rest_type)
zmt['rest_type'].value_counts()
```

```python
zmt.head()
```

```python
# cleaning location column

zmt['location'].value_counts()
```

```python
location = zmt["location"].value_counts(ascending = False)
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
# replacing lessthan 300 location to others

def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value
    
zmt['location'] = zmt['location'].apply(handle_location)
zmt['location'].value_counts()
```

```python
zmt.head()
```

```python
zmt['cuisines'].value_counts()
```

```python
cuisiens = zmt['cuisines'].value_counts(ascending=False)
cuisiens
```

```python
cuisiens_lessthan100 = cuisiens[cuisiens<100]
cuisiens_lessthan100
```

```python
# replacing lessthan100 cuisiens to others

def handle_cuisiens(value):
    if value in cuisiens_lessthan100:
        return 'others'
    else:
        return value
    
zmt['cuisines'] = zmt['cuisines'].apply(handle_cuisiens)
zmt['cuisines'].value_counts()
```
